```python
print(df1.duplicated().sum())
print(df2.duplicated().sum())
```

```python
print(df2.shape)
df2.drop_duplicates(inplace=True)
print(df2.shape)
```

```python
df =pd.merge(df1,df2,how= 'outer',on = 'PassengerId')
```

```python
df.columns= df.columns.str.lower()
```

```python
df
```

```python
df.isna().sum()
```

```python
df.isna().any(axis=1).sum()
```

```python
pd.pivot_table(df,index=['pclass','sex'],values=['age'],aggfunc='mean')
```

```python
df['tage']= round(df.groupby(['pclass','sex'])['age'].transform('mean'),0)
df['finalage']=np.where((df.age.isna()),df.tage,df.age)
```

```python
df.drop(columns= ['tage','age'],inplace = True)
```

```python
df.isna().sum()
```

```python
df['survival_tag']=np.where((df.survived==0),'NotSurvived','Survived')
```

```python
df.head()
```

```python
pd.pivot_table(df,index=['sex'],columns=['survival_tag'],values=['passengerid'],aggfunc='count',margins = True)
```

```python
pd.pivot_table(df,index=['sex'],values=['survived'],aggfunc='mean')
```

```python
df['qcut_age'] = pd.qcut(df.finalage,10)
df[['finalage','qcut_age']].sample(10)
```

```python
pd.pivot_table(df,index=['qcut_age'],columns=['sex'],values=['survived'],aggfunc='mean')
```

```python
df['passenger_type']= df.apply(lambda x : 'M' +str(x.pclass) if x.sex.lower() == 'male' else 'F'+str(x.pclass) , axis = 1)
df.head(5)
```

```python
pd.pivot_table(df,index=['passenger_type'],values=['survived'],aggfunc='mean')
```
